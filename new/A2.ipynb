## Cleaning and pre-processing

```python
print('Gender:', netflix['Gender'].unique())
print('Subscription Type:', netflix['Subscription Type'].unique())
print('Country:', netflix['Country'].unique())
print('Device:', netflix['Device'].unique())
```

## Converting text to binary

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
netflix['Gender'] = label_encoder.fit_transform(netflix['Gender'])
netflix.head(10)
```

```python
subscription = pd.get_dummies(netflix['Subscription Type'], dtype=int)
country = pd.get_dummies(netflix['Country'], dtype=int)
device = pd.get_dummies(netflix['Device'], dtype=int)
```

```python
netflix_clean = pd.concat([netflix, subscription, country, device], axis=1)
netflix_clean = netflix_clean.drop(['Subscription Type', 'Country', 'Device'], axis = 1) 
```

```python
netflix_clean.head()
```

```python
netflix_dropped_col = netflix_clean.drop(['User ID', 'Join Date', 'Last Payment Date', 'Plan Duration'], axis=1)
netflix_dropped_col
```

## Make sure the data is on the same scale

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
netflix_scaled = pd.DataFrame(scaler.fit_transform(netflix_dropped_col), columns = netflix_dropped_col.columns)
netflix_scaled.head(10)
```

## Check if all objects have been converted

```python
#checking if there're any column with of type 'object'
np.sum(netflix_scaled.dtypes == 'object')
```

## Write cleaned and pre-processed data to CSV file

```python
netflix_scaled.to_csv('netflix_prep.csv', index=False, encoding='utf-8')
```
