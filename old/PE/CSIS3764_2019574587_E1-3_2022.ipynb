```python
wine=wine[['Alcohol','Magnesium','Total_Phenols', 'Flavanoids',"Color_Intensity"]]
```

```python
wine.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
wine_scaled =pd.DataFrame(scaler.fit_transform(wine), columns=['Alcohol','Magnesium','Total_Phenols','Flavanoids','Color_Intensity'])
```

```python
wine_scaled
```

# Elbow curve

```python
import sklearn.cluster as cluster
import seaborn as sns

K = range(1,12)
wss = []
for k in K:
    kmeans=cluster.KMeans(n_clusters=k,init="k-means++")
    kmeans=kmeans.fit(wine_scaled)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)

mycenters = pd.DataFrame({'Clusters' : K, 'WSS' : wss})
sns.lineplot(x = 'Clusters', y = 'WSS', data = mycenters, marker="+")
```

# silhouette score

```python
import sklearn.metrics as metrics
SK = range(2,13)
sil_score = []
for i in SK:
    labels=cluster.KMeans(n_clusters=i,init="k-means++",random_state=200).fit(wine_scaled).labels_
    score = metrics.silhouette_score(wine_scaled,labels,metric="euclidean",sample_size=1000,random_state=200)
    sil_score.append(score)

sns.lineplot(x = 'Clusters', y = 'Sil Score', data = sil_centers, marker="+")
```

## Both The elbow curve and silhouette suggest that K = 3

```python
kmeans = cluster.KMeans(n_clusters=3 ,init="k-means++")
kmeans = kmeans.fit(wine)
```
